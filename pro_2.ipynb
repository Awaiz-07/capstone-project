```python
X = df[["Years_at_Company","Satisfaction_Level","Average_Monthly_Hours"]]
y = df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,mean_absolute_error
def results(predictions):
  print("m a e on m is{}",format(mean_absolute_error(Y_test,predictions)))
  print("r m s e on m is {}",format(np.sqrt(mean_squared_error(Y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
predictionlr=lr.predict(X_test)
```

```python
results(predictionlr)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rfmodel = RandomForestRegressor()
param_gridrfr={
    'n_estimators':[2,3,4,5,6],
    'max_depth':[5,10,15]
}
gridrfr = GridSearchCV(rfmodel,param_gridrfr)
gridrfr.fit(X_train,Y_train.values.ravel())
```
